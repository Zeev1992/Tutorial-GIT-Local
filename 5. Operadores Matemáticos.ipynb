__SUMA x + y__

```python
x = 6
y = 2
print(f"{x} mas {y} es igual a: {x+y}")
```

__RESTA X -Y__

```python
x = 8
y = 5
print(f"{x} menos {y} es igual a: {x-y}")
```

__MULTIPLICACIÓN x * y__

```python
x = 8
y = 5
print(f"{x} por {y} es igual a: {x*y}")
```

__DIVISIÓN x / y__

```python
x = 7
y = 2
print(f"{x} dividido por {y} es igual a: {x/y}")
```

__DIVISIÓN AL PISO. No consideran decimales (resto o residuo), solo se considera el COCIENTE, se utiliza doble barra //__

```python
z = 7
y = 2
print(f"{z} dividido al piso de {y} es igual a: {z//y}")
```

__MÓDULO. Son los decimales o lo que se conoce como *"resto o residuo"* en una división__


```python
z = 7
y = 2
print(f"{z} módulo de {y} es igual a: {z%y}")
```

__POTENCIA x ** y__

```python
x = 7
y = 2
print(f"{x} elevado a la potencia {y} es igual a: {x**y}")
```

```python
x = 7
print(f"{x} elevado a la potencia 3 es igual a: {x**3}")
```

__RAIZ CUADRADA {x ** 0.5}__

```python
x = 7
print(f"La raíz cuadrada de {x} es: {x**0.5}")
```

__RAÍZ CÚBICA {x ** (1./3.)}__

```python
y = 7
z = 2
x = y + z
print(f"La raíz cúbica de {x} es: {x**(1./3.)}")
```

```python
pow(2, 3)
```

Si se da un tercer argumento, pow(x,y,z), la función calcula primero "x" elevado a "y" y después calcula el resto de la devisión por z = x ** y = resultado % 5 = residuo o resto.

```python
pow(2, 3, 5)
```

##### Programa que pregunte su nombre y cuanto a vendido.
- El programa debe responder "Bien [nombre] ganaste $ 650 
- esta cantidad debe tener aplicado el 13%

```python
nombre = input("Cuál es tu nombre: ")
ventas = int(input("Cuánto has vendido en el mes? "))
comision = round(ventas * 13 / 100,2)
print(f"Muy bien {nombre}, tu comisión de este mes es: $ {comision}")
```
