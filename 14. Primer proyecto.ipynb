# Primer proyecto
* __Un código que pida el ingreso de un texto, párrafo, poema, etc__
* __Que pida ingresar tres letras__
    - Que diga cuantas veces aparece cada letra en el texto
    - Cuál es la letra con que inicia el texto
    - Cuál es la letra con que termina el texto
    - Que presente el texto en forma invertida
    - Qué diga si aparece la palabra 'Python' en el texto.

```python
texto = input("Ingresa un texto: ")
letras = input("Ingresa tres letras (sin espacios): ")

# Verificar que se ingresaron tres letras
if len(letras) != 3:
    print("Debes ingresar exactamente tres letras.")
    exit()

# Convertir las letras a minúsculas para hacer la búsqueda sin distinción de mayúsculas/minúsculas
letras = letras.lower()

contador = [0, 0, 0]  # Inicializar el contador para cada letra

# Recorrer cada letra en el texto
for letra in texto:
    letra_lower = letra.lower()  # Convertir la letra a minúscula

    if letra_lower in letras:
        # Obtener el índice de la letra encontrada en la cadena de letras ingresada
        index = letras.index(letra_lower)
        contador[index] += 1

# Mostrar los resultados para cada letra
for i in range(len(letras)):
    print("La letra", letras[i], "aparece", contador[i], "veces en el texto.")

# Obtener la letra primera y última del texto
letra_primera = texto[0]
letra_ultima = texto[-1]

print("La letra primera del texto es:", letra_primera)
print("La letra última del texto es:", letra_ultima)

# Invertir el orden de las palabras en el texto
palabras = texto.split()
texto_invertido = " ".join(palabras[::-1])

print("Texto invertido:", texto_invertido)

# Verificar si la palabra "Python" está presente en el texto
if "Python" in texto:
    print("La palabra 'Python' está presente en el texto.")
else:
    print("La palabra 'Python' no está presente en el texto.")
```

* __Otra forma:__

```python
# Pedir al usuario que ingrese un texto
texto = input("Ingresa un texto: ")

letras = []

# Convertimos el texto en minúsculas
texto = texto.lower()

# Pedir al usuario que ingrese las letras y que se agregue a 'letras'
letras.append(input("Ingresa la primera letra: ").lower())
letras.append(input("Ingresa la segunda letra: ").lower())
letras.append(input("Ingresa la tercer letra: ").lower())

# creamos un salto de línea
print("\n")

# Vamos a devolver la información al usuario
print("CANTIDAD DE LETRAS")

# Lo que hay en letras, la primer, segunda y tercer letra en el índice cero, 1 y 2 
cantidad_letras1 = texto.count(letras[0])
cantidad_letras2 = texto.count(letras[1])
cantidad_letras3 = texto.count(letras[2])

# Informamos al usuario
print(f"Hemos encontrado la letra '{letras[0]}' repetida {cantidad_letras1} veces")
print(f"Hemos encontrado la letra '{letras[1]}' repetida {cantidad_letras2} veces")
print(f"Hemos encontrado la letra '{letras[2]}' repetida {cantidad_letras3} veces")

# creamos un salto de línea
print("\n")

# Vamos a devolver la información al usuario
print("CANTIDAD DE PALABRAS")

# Creamos una variable 'palabras' que contenga esta información, lo que hay en texto, por medio de split separa en palabras
palabras = texto.split()
print(f"Hemos encontrado {len(palabras)} palabras ")

# creamos un salto de línea
print("\n")

# Vamos a devolver la información al usuario
print("LETRAS DE INICIO Y FIN")

letra_inicio = texto[0]
letra_final = texto[-1]
print(f"La letra inicial es '{letra_inicio}' y la letra final es '{letra_final}'")

# creamos un salto de línea
print("\n")

# Vamos a devolver la información al usuario
print("TEXTO INVERTIDO")
palabras.reverse()
texto_invertido = ' '.join(palabras)
print(f"Si ordenamos tu texto al revés va a decir: '{texto_invertido}'")

# creamos un salto de línea
print("\n")

# Vamos a devolver la información al usuario
print("BUSCANDO LA PALABRA PYTHON")
buscar_python = 'python' in texto
dic = {True:"si", False:"no"}
print(f"La palabra 'Python' {dic[buscar_python]} se encuentra en el texto")
```
