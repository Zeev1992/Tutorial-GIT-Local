# Ejercicio 1. Crear un repositorio local
1. Ahora vamos a crear un nuevo directorio para nuestro repositorio local.
Cree un myrepo directorio copiando y pegando el __mkdir__ siguiente comando en la terminal:
## mkdir mirepo
2. Vaya al __mirepo__ copiando y pegando el __cd__ siguiente comando:
### cd mirepo
3. En este __mirepo__, creemos un nuevo repositorio local de git usando el __git init__ comando. Copie y pegue el siguiente comando en la terminal:
### git init
4. __Ahora se crea un nuevo repositorio local__, que puede verificar haciendo una lista de directorios pegando el siguiente comando en la ventana de la terminal:
### ls -la .git
El resultado muestra el contenido del __.git__ subdirectorio que alberga el __repositorio local__:


# Ejercicio 2: crear y agregar un archivo al repositorio local
1. Ahora vamos a crear un archivo vacío usando el siguiente __touch__ <nombre de archivo> comando:
### touch nuevoarchivo
2. Agregue este archivo al repositorio usando el siguiente __git add__ comando:
### git add nuevoarchivo

# Ejercicio 3: Confirmar cambios
1. Antes de que podamos confirmar nuestros cambios, debemos decirle a __git__ quiénes somos. Podemos hacer esto usando los siguientes comandos:
### git config --global user.email "you@example.com"
### git config --global user.name "Your Name"
#### __NOTA:__ Debes escibri tu correo y user name con el que te registraste en Github
2. Una vez que el repositorio tenga el archivo creado, confirmemos nuestros cambios usando el siguiente __git commit__ comando. Tenga en cuenta que la confirmación requiere un mensaje que incluimos usando el __-m__ parámetro:
### git commit -m "added nuevoarchivo"


# Ejercicio 4: Crear una rama
NOTA: Nuestra confirmación anterior creó una rama principal predeterminada llamada __master__.
1. Para realizar cambios posteriores en nuestro repositorio, creemos una nueva rama en nuestro repositorio local. Copie y pegue el siguiente __git branch__ comando en la terminal para crear una rama llamada __mirabranch__:
### git branch mirabranch

### git checkout mirabranch
2. Verifiquemos que la nueva rama ahora sea la rama activa emitiendo el siguiente __git branch__ comando:
### git branch
3. Tenga en cuenta que el asterisco __*__ ahora está al lado de lo __mirabranch__ que indica que ahora está activo:

#### NOTA: Como acceso directo para crear y usar la rama git branch y luego activarla, git checkout puede usar el acceso directo de la siguiente manera con la -b opción que crea la rama y la activa en un solo paso:
### git checkout -b mirabranch

# Ejercicio 9: Revertir cambios confirmados
1. A veces, es posible que no pruebe completamente los cambios antes de realizarlos y puede tener consecuencias no deseadas... __puede revertir sus cambios__ usando un __git revert__ comando como el siguiente.
